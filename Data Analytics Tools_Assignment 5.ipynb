```python
#Import essesntial libraries

import numpy
import matplotlib
import matplotlib.pyplot as plt
```

## Creating Required Functions

### Visualization Function

```python
# Creating Visualization function

def visualize(filename):
    """This function will visualize Average, Max and Min plots"""

    data = numpy.loadtxt(fname=filename, delimiter=',')

    fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

    axes1 = fig.add_subplot(1, 3, 1)
    axes1.set_ylabel('Values')
    axes1.set_xlabel('Feature#')
    axes1.plot(numpy.mean(data, axis=0))
    axes1.plot(numpy.max(data, axis=0))
    axes1.plot(numpy.min(data, axis=0))
    plt.legend(['Avg','Max','Min'])
    
    fig.tight_layout()
    matplotlib.pyplot.show()
```

### Issues Identification Function

```python
# Creating Identify issues function

def identify_issues(filename):
    """This function will identify issues basis conditions"""

    data = numpy.loadtxt(fname=filename, delimiter=',')

    if numpy.max(data, axis=0)[0] == 0 and numpy.max(data, axis=0)[20] == 20:
        print('I doubt the integrity of this dataset\n')
    
    elif sum(numpy.min(data,axis=0)) == 0:
        print('Why is this happening\n')
    
    else:
        print('Dataset looks good\n')
```

## Iterating over datasets using glob()

```python
import glob
import numpy 
import matplotlib
import matplotlib.pyplot 

dataset = sorted(glob.glob('data*.csv'))

for datasets in dataset[:4]:
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
